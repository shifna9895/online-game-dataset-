# Filter and sort

## Filter games released in a specific year

```python
releasing_year=data[data['Year']==2015].reset_index(drop=True)
# releasing_year.reset_index(drop=True,inplace=True)
releasing_year
```

## filter and sort the vgsales_online_game dataset to find all games released in 2014, displaying their name, platform, publisher, and total sales ratio, sorted by platform in ascending order and total sales ratio in descending ?

```python
releasedIn2014=data.loc[data['Year']==2014][['Game_Name','Platform','Publisher','Total_sales_ratio']].sort_values(by=['Platform','Total_sales_ratio'],ascending=[True,False])
releasedIn2014.reset_index(drop=True,inplace=True)
releasedIn2014
```

# groupby()

## group the dataset by both game name and year, and then calculate the maximum total sales ratio and minimum global sales for each game in each year?

```python
gr=data.groupby(['Game_Name','Year']).agg({'Total_sales_ratio':max,'Global_Sales':min})
gr
```

## group the dataset by game name and platform, and then calculate the maximum, minimum, mean, and total North American sales for each combination?

```python
group=data.groupby(['Game_Name','Platform'])['NA_Sales'].agg(['max','min','mean','sum'])
group.head()
```
